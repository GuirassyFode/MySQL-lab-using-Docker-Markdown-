# Database Administration and SQL Language

#         Relational Database Management Systems (RDBMS) 

**About RDBMS**

   • This refers to the software system that is used to 
interact and maintain a relational database.

   • It stores data in a structured format using columns 
and rows.

   • The values within a table are related to each other.

   • Tables can be related to other tables.

   • It implements SQL (Structured Query Language) 
for querying and managing data within a 
relational database.
   
   • RDMS supports the use of stored procedures 
within the database.
About Keys
   
   • Keys are used in order to uniquely identify rows 
within a table (i.e., a primary key).
   
   • They can be used to form a link between tables 
(i.e., a foreign key)
